#### 2. Define and differentiate the following terms:

- Brightness
- Contrast
- Dynamic Range

**Brightness:** This refers to the overall lightness or darkness of an image. Adjusting brightness uniformly increases or decreases all the pixels' light levels, making the entire image lighter or darker.

**Contrast:** Contrast is the difference between the darkest and lightest parts of an image. High contrast means a greater difference, resulting in a more vivid image, while low contrast means a smaller difference, producing a flatter image.

**Dynamic Range:** Dynamic range is the measure of the range of brightness levels an image can display, from the darkest shadows to the brightest highlights. A higher dynamic range captures more details in both very dark and very bright areas, leading to a richer and more detailed image.

- **Brightness** affects the overall lightness or darkness of an image.
- **Contrast** affects the difference between the darkest and lightest parts.
- **Dynamic Range** refers to the range of brightness levels an image can display, affecting the detail in shadows and highlights.

**Histogram Equalization and its Role in Image Enhancement**

Histogram Equalization is a technique used to improve the contrast of an image by redistributing the intensities of the pixels. Essentially, it spreads out the most frequent intensity values, which enhances the overall contrast of the image. This technique is particularly useful in situations where the image's lighting is not uniform, such as overexposure or underexposure.

Steps:

**Calculate the Histogram:** Determine the frequency of each intensity level in the image.

**Compute the Cumulative Distribution Function (CDF):** This gives us a cumulative sum of the histogram values.

**Normalize the CDF:** Scale it to the range of intensity levels (typically 0 to 255 for an 8-bit image).

**Map the Original Intensities to the New Intensities:** Using the normalized CDF, remap the original pixel intensities to the new intensities.

This process ensures that the output image has a more uniform distribution of intensities, which enhances the overall visual contrast.

By applying this transformation, the histogram of the resulting image is more evenly distributed, leading to an image with improved contrast.

**Spatial Filtering vs. Frequency Domain Filtering**

**Spatial Filtering:**

- Involves processing the image directly in the spatial domain (the domain of pixel values).

- Uses a filter or kernel (such as a Gaussian blur or Sobel operator) that is applied to the pixels of the image to produce a new image.

- Typically used for tasks like smoothing, sharpening, and edge detection.

- Example: Applying a 3x3 averaging filter to an image to reduce noise.

**Frequency Domain Filtering:**

- Involves processing the image in the frequency domain by transforming it using the Fourier Transform.

- The image is represented in terms of its frequency components, allowing for manipulation of these frequencies.

- Typically used for tasks like filtering out specific frequency ranges (e.g., noise removal).

- Example: Applying a low-pass filter in the frequency domain to reduce high-frequency noise.

**Low-Pass Filters vs. High-Pass Filters**

**Low-Pass Filters:**

- Allow low-frequency components to pass through while attenuating (reducing) high-frequency components.

- Used to smooth images by reducing fine details and noise.

- Example: Gaussian Blur, which removes high-frequency noise and smooths the image.

**High-Pass Filters:**

- Allow high-frequency components to pass through while attenuating low-frequency components.

- Used to enhance edges and fine details in an image.

- Example: Sobel Operator, which detects edges by highlighting areas with rapid intensity changes.



Spatial Filtering operates directly on pixel values, while Frequency Domain Filtering works on the image's frequency components.

Low-Pass Filters smooth images by reducing high-frequency noise, while High-Pass Filters enhance edges by emphasizing high-frequency details.

## Implementation

**1 .Brightness and Contrast Adjustment**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Load the image
image_path = "/kaggle/input/bird-image/bird.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Converted Into Grayscale image (single channel)
# Apply brightness and contrast adjustment
alpha = 1.2 # Adjust Contrast, An alpha greater > 1 increases the contrast, while < 1 decreases it.
beta = 50 # Brightness adjustment, A positive beta increases the brightness, while a negative beta decreases it.
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
# Display Original and Adjusted Images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(adjusted_image, cmap="gray")
plt.title("Brightness & Contrast Adjusted")
plt.axis("off")

plt.tight_layout()
plt.show()
```

### 2. Histogram Equalization

```python

image_path = "/kaggle/input/bird-image/bird.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)
# Display Histogram Equalized Image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap="gray")
plt.title("Histogram Equalized Image")
plt.axis("off")
plt.tight_layout()
plt.show()
```

### 3. Smoothing Filters

```python

# Dataset
image_path = "/kaggle/input/bird-image/bird.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


# Apply Smoothing Filters
average_filter = cv2.blur(image, (5, 5))
gaussian_filter = cv2.GaussianBlur(image, (5, 5), 0)
median_filter = cv2.medianBlur(image, 5)
# Display Filtered Images
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(average_filter, cmap="gray")
plt.title("Average Filter")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(gaussian_filter, cmap="gray")
plt.title("Gaussian Filter")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(median_filter, cmap="gray")
plt.title("Median Filter")
plt.axis("off")
plt.tight_layout()
plt.show()
```

## 4. Edge Detection

```python

# Dataset 
image_path = "/kaggle/input/bird-image/bird.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


# Sobel, Prewitt, and Laplacian Filters
sobel_edges = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
prewitt_kernel = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_edges = cv2.filter2D(image, -1, prewitt_kernel)
laplacian_edges = cv2.Laplacian(image, cv2.CV_64F)
# Display Edge Detection Results
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(sobel_edges, cmap="gray")
plt.title("Sobel Edge Detection")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(prewitt_edges, cmap="gray")
plt.title("Prewitt Edge Detection")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(laplacian_edges, cmap="gray")
plt.title("Laplacian Edge Detection")
plt.axis("off")
plt.tight_layout()
plt.show()
```

### 5. Sharpening Filter

```python
# Apply Laplacian-based Sharpening
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_image = cv2.filter2D(image, -1, kernel)
# Display Sharpened Image
plt.figure(figsize=(8, 6))
plt.imshow(sharpened_image, cmap="gray")
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```
