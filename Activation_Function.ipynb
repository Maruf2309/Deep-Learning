# 1. ReLU - Rectified Linear Unit (Hidden Layer)


```python
def relu(x):
  return np.maximum(0, x)
```

```python
print('Positive Value :',relu(10)) # for positive return values as it it
print('Negative Value :',relu(-10))  # for negaive its takes
```

```python
# List of values for verificaiton
relu(np.array([-10, -15, 0, 1, 4, 10, 15])) # for negative its returned 0 & positive as it is!
```

# 2. Leaky ReLU (Hidden Layer)


Leaky ReLU allows a small, positive gradient (alpha) when the unit is not active.

- `Alpha`: is  very small positive number here we define as 0.01 which will be multplied when there is only `Negative Values`


```python
def leaky_relu(x, alpha = 0.01):
  return np.where(x > 0, x, alpha * x)   # where x > 0, return as it is (x), otherewise multiply with alpha
```

```python
leaky_relu(-15)
```

```python
# How works, Verified
-15 * 0.01
```

```python
leaky_relu(np.array([-10, -15, 0, 1, 4, 10, 15])) # Positive values as it is, negives are multiplied with given Alpha value
```

# 3. ELU - Exponential Linear Unit (Hidden Layer)



ELU has a small slope for negative values, unlike ReLU, to avoid the dying ReLU problem.


`X :` When >=0 then X, if `< 0 /negative` then `Nag value * aplha with **exponential**`

```python
def elu(x, alpha = 1.0):
  return np.where(x > 0, x, alpha *(np.exp(x) - 1))
```

```python
print('Positive Value : ',elu(5.0))
```

```python
elu(np.array([-1, 0, 1, 2]))
```

# 4. Sigmoid (Output Layer & Binaryclass Problem)

Its also known as Logistic Function, range (0,1), used in output layer only. Regardless the positive / negative it will convert negaive to 0 and max positive number into 1. Returned only `Probabilistic` Values only

`Disadvantage`: Vanishing Gradient, Non Zero-Centerd output, Higher Computational Complexity

```python
def sigmoid(x):
  return 1 / (1+ np.exp(x))
```

```python
sigmoid(np.array([-1, 0, 1, 2]))
```

# 5. Softmax (Output Layer & Multiclass Problem)


Softmax converts a real vector to a vector of categorical probabilities.



image.png

```python
def softmax(x):
  e_x = np.exp (x - np.max(x))
  return e_x / e_x.sum(axis = 0)
```

```python
proba = softmax(np.array([2, 1, .5]))
proba
```

```python
sum(proba) # Alwyas 1
```

```python
np.argmax(proba) # Return the index values for which its returned the highest probality score, here 0 index has the max probality
```

# 6. Swish (Hidden Layer)


Swish is a self-gated activation function.



```python
def swish(x, beta=1.0):
    return x * sigmoid(beta * x)
```

```python
swish(5)
```

```python
swish(np.array([-1, 0, 1, 2]))
```
